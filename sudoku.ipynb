```python
# 수정본 1

import pygame
import sys
import time

pygame.init()
WIDTH, HEIGHT = 540, 650  
SCREEN = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("나만의 커스텀 스도쿠")


WHITE = (255, 255, 255)
BLACK = (30, 30, 30)      
SOFT_BLUE = (70, 130, 180) 
GRAY = (200, 200, 200)
LINE_COLOR = (50, 50, 50)
RED = (220, 20, 60)


FONT = pygame.font.SysFont("malgungothic", 35)
TIMER_FONT = pygame.font.SysFont("malgungothic", 40)


initial_board = [
    [7, 8, 0, 4, 0, 0, 1, 2, 0],
    [6, 0, 0, 0, 7, 5, 0, 0, 9],
    [0, 0, 0, 6, 0, 1, 0, 7, 8],
    [0, 0, 7, 0, 4, 0, 2, 6, 0],
    [0, 0, 1, 0, 5, 0, 9, 3, 0],
    [9, 0, 4, 0, 6, 0, 0, 0, 5],
    [0, 7, 0, 3, 0, 0, 0, 1, 2],
    [1, 2, 0, 0, 0, 7, 4, 0, 0],
    [0, 4, 9, 2, 0, 6, 0, 0, 7]
]
board = [row[:] for row in initial_board]
original_mask = [[board[r][c] != 0 for c in range(9)] for r in range(9)]

selected = None
start_time = time.time()

def is_valid(b, num, pos):
    for i in range(9):
        if b[pos[0]][i] == num and pos[1] != i: return False
        if b[i][pos[1]] == num and pos[0] != i: return False
    bx, by = pos[1] // 3, pos[0] // 3
    for i in range(by*3, by*3+3):
        for j in range(bx*3, bx*3+3):
            if b[i][j] == num and (i, j) != pos: return False
    return True

def draw_window():
    SCREEN.fill(WHITE)
    
    
    elapsed = int(time.time() - start_time)
    min, sec = divmod(elapsed, 60)
    
    
    time_str = f"{min:02d}:{sec:02d}"
    
    timer_text = TIMER_FONT.render(time_str, True, BLACK, WHITE)
    
    SCREEN.blit(timer_text, (20, 5))

    offset_y = 50
    for i in range(10):
        thick = 4 if i % 3 == 0 else 1
        pygame.draw.line(SCREEN, LINE_COLOR, (i * 60, offset_y), (i * 60, 540 + offset_y), thick)
        pygame.draw.line(SCREEN, LINE_COLOR, (0, i * 60 + offset_y), (540, i * 60 + offset_y), thick)

    num_counts = {i: 0 for i in range(1, 10)}
    for r in range(9):
        for c in range(9):
            val = board[r][c]
            if val != 0:
                num_counts[val] += 1
                color = BLACK if original_mask[r][c] else SOFT_BLUE
                
                if not is_valid(board, val, (r, c)): color = RED
                
                txt = FONT.render(str(val), True, color)
                SCREEN.blit(txt, (c * 60 + 22, r * 60 + offset_y + 10))

    pad_y = 650
    display_nums = [str(n) for n, count in num_counts.items() if count < 9]
    pad_text = FONT.render("  ".join(display_nums), True, BLACK)
    SCREEN.blit(pad_text, (WIDTH // 2 - pad_text.get_width() // 2, pad_y - 40))

    if selected:
        pygame.draw.rect(SCREEN, SOFT_BLUE, (selected[1] * 60, selected[0] * 60 + offset_y, 60, 60), 3)

    pygame.display.update()

def main():
    global selected
    run = True
    while run:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                run = False
            if event.type == pygame.MOUSEBUTTONDOWN:
                x, y = pygame.mouse.get_pos()
                if 50 <= y <= 590:
                    selected = ((y - 50) // 60, x // 60)
            if event.type == pygame.KEYDOWN and selected:
                r, c = selected
                if not original_mask[r][c]: 
                    if pygame.K_1 <= event.key <= pygame.K_9:
                        board[r][c] = event.key - pygame.K_0
                    elif event.key in [pygame.K_BACKSPACE, pygame.K_0, pygame.K_DELETE]:
                        board[r][c] = 0
        draw_window()
    pygame.quit()

if __name__ == "__main__":
    main()
```
